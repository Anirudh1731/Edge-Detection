# **HOUGH LINE TRANSFORM ***

# 1)First of all edge is to be detected using any edge detector 
# 2)Mapping the points in the edge into the hough space
# the edges are stored in accumulator 
# 3)then the points in the accumulator  are  interpreted to make lines of infinite length
# 4)Converting infinite length into finite one 

# BASICALLY IN THIS A POINT IS XY OR POLAR PLANE IS CONVERTED INTO M-C SPACE
 

# In hough line tranform we are basically mapping each points in the image to pints in the m-c space or r-theta space for each each r,theta in m-c space a minimum vote is given or threshold is given which acts as a deciding factor for being declared as line. As the value increases in the accumulator it can be easily decided as a line.
# the function HOUGH TRANSORM RETURNS the set of r,0 value.

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread('/content/sudoku_board_fig1.jpg')
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges=cv2.Canny(gray,50,150,apertureSize=3)
lines=cv2.HoughLinesP(edges,1,np.pi/180,100,minLineLength=100,maxLineGap=10)
print(lines)
# (img,rho,theta,threshold)
cv2_imshow(edges)
for line in lines:
  x1,y1,x2,y2=line[0] #here we are defining the initial and final point for a line
  # because the lines is an array and there is only one element in each sub array
  cv2.line(img,(x1,y1),(x2,y2),(255,0,0),2)
  #(image,start,end.color,thickness)
cv2_imshow(img)
# k=cv2.waitkey(0)
# cv2.destroyAllWindows
```
