# **Importing packages**

```python
import cv2
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img=cv2.imread("/content/spiderman.jpg")
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
image_sobel_x=cv2.Sobel(gray,-1,1,0)
image_sobel_y=cv2.Sobel(gray,-1,0,1)
image_sobel_xy=cv2.addWeighted(image_sobel_x,0.5,image_sobel_y,0.5,0)
image_laplacian=cv2.Laplacian(gray,-1)

canny=cv2.Canny(img,100,200)
cv2_imshow(canny)
cv2_imshow(image_sobel_x)
cv2_imshow(image_sobel_y)
cv2_imshow(image_sobel_xy)
cv2_imshow(image_laplacian)


```

```python
titles=['image','canny']
images=[img,canny]
for i in range(2):
  plt.subplot(2,1,i+1),plt.imshow(images[i],'gray')
  plt.title(titles[i])
  plt.xticks([]),plt.yticks([])
plt.show()  
```
